# Eddy Import and Selection for Southern Ocean 

```python
import intake
import numpy as np
import matplotlib.pylab as plt
import xarray as xr
import matplotlib.colors as colors
```

```python
#importing data

eerie_cat=intake.open_catalog("https://raw.githubusercontent.com/eerie-project/intake_catalogues/main/eerie.yaml")
da_atmos = eerie_cat["dkrz"]["disk"]["model-output"]["icon-esm-er"]["hist-1950"]["v20240618"]["atmos"]["gr025"]["2d_daily_mean"].to_dask()
```

```python
#importing eddies

ed_ac = xr.open_dataset("/work/bk1377/b382618/model_eddy_eval/historical/icon/eddyTracks/anticyclonic_tracks.nc", chunks = {"obs": 10000})
ed_c = xr.open_dataset("/work/bk1377/b382618/model_eddy_eval/historical/icon/eddyTracks/cyclonic_tracks.nc", chunks = {"obs": 10000})
```

```python
#selecting for eddies in southern ocean

edso_ac = ed_ac.where(ed_ac.latitude.compute() < -60, drop = True)
edso_c = ed_c.where(ed_c.latitude.compute() < -60, drop = True)
```
